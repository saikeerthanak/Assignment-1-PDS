```python
# Bar Chart of Gender vs. Average Score

import pandas as pd
import matplotlib.pyplot as plt

# Load your dataset 
data = pd.read_csv('C:\Desktop\PDS\Assignment 1\Student Performance\Data Preprocessing\cleaned_data.csv')

# Assuming your dataset has columns 'race/ethnicity' and 'Average Score'
gender = data['gender']
average_scores = data['average score']

# Create a bar chart
plt.figure(figsize=(10, 6))
data_grouped = data.groupby('gender')['average score'].mean()
data_grouped.plot(kind='bar', color='skyblue')
plt.xlabel('Gender')
plt.ylabel('Average Score')
plt.title('Bar Chart of Gender vs. Average Score')
plt.xticks(rotation=0)

# Save the histogram
plt.savefig('visualizations\\bar_chart.png', bbox_inches='tight')

# Show the histogram
plt.show()
```
