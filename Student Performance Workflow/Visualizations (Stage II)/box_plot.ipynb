```python
# Box Plot of Parental Level of Education vs. Average Score

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load your dataset 
data = pd.read_csv('C:\Desktop\PDS\Assignment 1\Student Performance\Data Preprocessing\cleaned_data.csv')

# Assuming your dataset has columns 'parental level of education' and 'Average Score'
plt.figure(figsize=(12, 6))
sns.boxplot(x='parental level of education', y='average score', data=data, palette='Set3')
plt.xlabel('Parental Level of Education')
plt.ylabel('Average Score')
plt.title('Box Plot of Parental Level of Education vs. Average Score')
plt.xticks(rotation=45, horizontalalignment='right')

# Save the histogram
plt.savefig('visualizations\\box_plot.png', bbox_inches='tight')
# Show the histogram
plt.show()
```
