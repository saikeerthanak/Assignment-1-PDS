```python
# Correlation Matrix Heatmap of Lunch vs. Average Score

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load your dataset 
data = pd.read_csv('C:\Desktop\PDS\Assignment 1\Student Performance\Data Preprocessing\cleaned_data.csv')

# Assuming your dataset has columns 'lunch' and 'Average Score'
data['lunch'] = data['lunch'].map({'standard': 1, 'free/reduced': 0})  # Convert to numeric

# Calculate the correlation matrix
correlation_matrix = data[['lunch', 'average score']].corr()

# Create a heatmap of the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix Heatmap of Lunch vs. Average Score')

# Save the histogram
plt.savefig('visualizations\correlation_matrix_heatmap.png', bbox_inches='tight')
# Show the histogram
plt.show()
```
