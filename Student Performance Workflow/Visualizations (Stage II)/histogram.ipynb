```python
#Histogram of Test Preparation vs. Average Score

import pandas as pd
import matplotlib.pyplot as plt

# Load your dataset 
data = pd.read_csv('C:\Desktop\PDS\Assignment 1\Student Performance\Data Preprocessing\cleaned_data.csv')

# Assuming your dataset has columns 'Test Preparation Course' and 'Average Score'
test_prep_scores = data['test preparation course']
average_scores = data['average score']

# Create a figure and axis for the histogram
fig, ax = plt.subplots(figsize=(8, 6))


# Create a histogram of 'Average Score' for each category of 'Test Preparation Course'
plt.hist([average_scores[test_prep_scores == 'completed'], average_scores[test_prep_scores == 'none']], 
         bins=10, alpha=0.5, color=['blue', 'red'], label=['Completed', 'None'])

# Customize the plot
plt.xlabel('Average Score')
plt.ylabel('Frequency')
plt.title('Histogram of Test Preparation vs. Average Score')
plt.legend()

# Save the histogram
plt.savefig('visualizations\histogram.png', bbox_inches='tight')
# Show the histogram
plt.show()
```
