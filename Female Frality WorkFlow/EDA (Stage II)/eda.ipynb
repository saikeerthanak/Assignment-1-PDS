```python
# eda.ipynb

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the cleaned data
cleaned_data = pd.read_csv("C:\Desktop\PDS\Assignment 1\Female Frality Workflow\Data processing/cleaned_data.csv")
# Descriptive statistics
print(cleaned_data.describe())

# Visualization 1: Histogram of Age
plt.figure(figsize=(8, 6))
sns.histplot(cleaned_data['Age'], bins=10, kde=True)
plt.title("Histogram of Age")
plt.savefig("visualizations/histogram_age.png")
plt.show()

# Visualization 2: Scatter plot of Grip strength vs. Age
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Age', y='Grip strength (kg)', data=cleaned_data)
plt.title("Scatter plot of Grip strength vs. Age")
plt.savefig("visualizations/scatter_age_grip.png")
plt.show()
```
